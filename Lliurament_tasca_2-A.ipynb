### a) Carregareu qualsevol imatge (jpg, png ..) amb Matplotlib. adoneu-vos que les imatges RGB (Red, Green, Blue) són realment només amplades × alçades × 3 matrius (tres canals Vermell, Verd i Blau), una per cada color de nombres enters int8, manipuleu aquests bytes i torneu a utilitzar Matplotlib per desar la imatge modificada un cop hagueu acabat.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('img_forest.jpg')
plt.imshow(img)
```

Veiem que la imatge és amplada x alçada x 3:

```python
print(img)
```

A continuació crearem una funció que apliqui manualment el filtre _grayscale_, o escala de grisos. Tenint en compte que la escala de grisos succeeix quan els 3 valors RGB són iguals, utilitzarem la mitjana dels 3 valors per a establir el to de gris que tindrà cada píxel.

```python
img.shape
```

```python
def grayscale(img):
    grayscaled_img = img.copy()
    for x in grayscaled_img:
        for y in x:
            # El valor mitjà dels tres canals serà el nou valor
            y[0] = y[1] = y[2] = np.mean(y)
    return grayscaled_img
```

```python
grayscale(img)
```

```python
plt.imshow(grayscale(img))
```

Guardarem la imatge editada:

```python
plt.imsave('img_forest_grayscale.jpg', grayscale(img))
```
